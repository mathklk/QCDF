```python
tests = [
    Scenario('ULA', 3.5, ula3_5),
    Scenario('ULA', 20,  ula20 ),
    Scenario('ULA', 100, ula100),
    Scenario('UCA', 20,  uca20 )
]

for i, scenario in enumerate(tests):
    scenario.df = pd.read_csv(io.StringIO(scenario.data), sep="\t", index_col=0)

# angles180 = np.arange(-180, 181, 30)
# angles90 = np.arange(-90, 91, 30)
```

```python
angles90  = tests[ 0].df.index.unique()
angles180 = tests[-1].df.index.unique()
rmse_report = pd.DataFrame(index=[str(t) for t in tests], columns=angles180, dtype=float)
std_report = rmse_report.copy()
for test in tests:
    scope = angles90 if test.type == 'ULA' else angles180
    for true_angle in scope:
        true_angles = []
        measured_angles = []
        batch = test.df.loc[test.df.index.isin([true_angle])]
        for i in range(len(batch)):
            spectrum = batch.iloc[i]
            true_angles.append(true_angle)
            measured_angles.append(spectrum.idxmax())

        true_arr = np.asarray(pd.to_numeric(true_angles), dtype=float)
        measured_arr = np.asarray(pd.to_numeric(measured_angles), dtype=float)

        rmse = root_mean_squared_error(true_arr, measured_arr)
        rmse_report.loc[str(test), true_angle] = rmse

        std = np.std(measured_arr - true_arr)
        std_report.loc[str(test), true_angle] = std

print('RMSE')
print('\t'.join(rmse_report.columns.astype(str)))
print(rmse_report.to_csv(sep="\t", index=False, header=False, decimal=','))
print()
print('STD')
print('\t'.join(std_report.columns.astype(str)))
print(std_report.to_csv(sep="\t", index=False, header=False, decimal=','))
```

```python
def plotScenario(scenario: Scenario, ax: Axes):
    # fig, ax = plt.subplots(figsize=FIG_SIZE)
    ax.set_title("MUSIC " + scenario.name())
    df = scenario.df
    angles = df.index.unique().astype(float)
    frequencies = df.columns.astype(float)

    # Z_norm = (df.values - df.values.min(axis=1, keepdims=True)) / (df.values.max(axis=1, keepdims=True) - df.values.min(axis=1, keepdims=True))

    for i, angle in enumerate(angles):
        y_offset = i
        overlap = .7
        alpha = .7
        # z_norm jus for this angle
        Z_norm = (df.loc[angle].values - df.loc[angle].values.min()) / (df.loc[angle].values.max() - df.loc[angle].values.min())

        ax.vlines(
            min(angle, frequencies.max()),
            y_offset, 
            y_offset + overlap, 
            colors='black', 
            linestyles='--', 
            linewidth=1.5,
            zorder=101-i,
            label='Wahrer Winkel'#'$\\alpha$'
        )

        for _, series in df.loc[angle].iterrows():
            spectrum = series.to_numpy()
            Z_norm = (spectrum - spectrum.min()) / (spectrum.max() - spectrum.min())
            ax.fill_between(
                frequencies, 
                y_offset, 
                y_offset + Z_norm * overlap, 
                alpha=alpha, 
                color=scenario.color(), 
                edgecolor='darkgray',#scenario.color(), 
                linewidth=1, 
                zorder=100-i,
                label='MUSIC Spektrum'#$\\text{P}(\\alpha)$'
            )

            peak_idx = np.argmax(spectrum)
            peak_freq = frequencies[peak_idx]
            ax.scatter(
                peak_freq, 
                y_offset + Z_norm[peak_idx] * overlap,
                color=scenario.color(),
                alpha=alpha,
                marker='D',
                s=20,
                zorder=102-i,
                label='Maximum'#'$\\max\\text{P}(\\alpha)$'
            ).set_path_effects([
                pe.Stroke(linewidth=3, foreground='black'), 
                pe.Normal()
            ])

        # ).set_path_effects([
        #     pe.Stroke(linewidth=3, foreground='white'), 
        #     pe.Normal()
        # ])
        
        # angle = min(frequencies.max(), max(frequencies.min(), angle))
        # ax.scatter(
        #     angle, 
        #     y_offset,#y_offset + spectrum[frequencies.get_loc(angle)],
        #     color='red', 
        #     s=50, 
        #     zorder=102-i
        # )

    yticks = range(0, len(angles), max(1, len(angles)//10))
    ax.set_yticks(yticks)
    ax.set_yticklabels([str(int(angles[i])) for i in yticks])
    xticks = list(angles)
    if scenario.type == 'UCA':
        xticks = [-180] + xticks
    ax.set_xticks(xticks)

    ax.set_xlabel('Pseudospektrum Winkel in °')
    # ax.set_ylabel('$\\alpha$ in °')
    ax.set_ylabel('Wahrer Winkel in °')

    # ax.set_title(f'MUSIC Summen Spektrum - {scenario.name()}')
    # ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)

    # _min, _max = df.values.min(), df.values.max()
    # print(f"Scenario {scenario.name()} min: {_min}, max: {_max}")

ula_fig, ula_axs = subplots(nrows=3)
# set vspace higher
ula_fig.subplots_adjust(hspace=0.3)

for (scenario, ax) in zip([test for test in tests if test.type == 'ULA'], ula_axs):
    plotScenario(scenario, ax)

uca_fig, uca_ax = plt.subplots(figsize=FIG_SIZE*(1, 1.2))
plotScenario([test for test in tests if test.type == 'UCA'][0], uca_ax)

# remove duplicate handles
for ax in list(ula_axs) + [uca_ax]:
    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1, 1))


ula_fig.savefig('plots/MUSIC_Separate_Ridges_ULA.pdf', bbox_inches='tight')
uca_fig.savefig('plots/MUSIC_Separate_Ridges_UCA.pdf', bbox_inches='tight')
```
