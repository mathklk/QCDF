```python
import pandas as pd
import numpy as np
import io
from dataclasses import dataclass
from typing import *
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# load previous cell as text

@dataclass
class Scenario:
    type: str
    distance: float
    data: str
    color: str

    df: pd.DataFrame = None # type: ignore

    def name(self):
        return f"{self.type} {self.distance}m"



tests = [
    Scenario('ULA', 3.5, ula3_5, "#004c6d"),
    Scenario('ULA', 20,  ula20 , "#6698b3"),
    Scenario('ULA', 100, ula100, '#beebff'),
    Scenario('UCA', 20,  uca20 , '#ed7d31')
]

# String Copy/Paste interpretieren

for i, scenario in enumerate(tests):
    df = pd.read_csv(io.StringIO(scenario.data), sep="\t", header=None)
    df.index = [
        "Anzahl Peilungen Insgesamt",
        "Anzahl Peilungen mit Antwort",
        "MUSIC Peaks Separat - Mittelwert",
        "MUSIC Peaks Separat - Median",
        "MUSIC Peaks Separat - Standardabweichung",
        "MUSIC Peaks Separat - RMSE",
        "MUSIC Summe - Peak",
        "MUSIC Summe - Qualität",
        "PDOA01 - Mittelwert",
        "PDOA01 - Median",
        "PDOA01 - Standardabweichung",
        "PDOA01 - RMSE",
        "PDOA12 - Mittelwert",
        "PDOA12 - Median",
        "PDOA12 - Standardabweichung",
        "PDOA12 - RMSE",
        "PDOA02 - Mittelwert",
        "PDOA02 - Median",
        "PDOA02 - Standardabweichung",
        "PDOA02 - RMSE",
        "PDOA Mittel - Mittelwert",
        "PDOA Mittel - Median",
        "PDOA Mittel - Standardabweichung",
        "PDOA Mittel - RMSE"
    ]
    if scenario.type == 'ULA':
        df.columns = [-90, -60, -30, 0, 30, 60, 90]
    else:
        df.columns = [-150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180]
        # Kopiere 180° zu -180° für übergang in der Darstellung
        df.insert(0, -180, df.iloc[:, -1])
        

    tests[i].df = df

# Automatische, "Statische" Berechnungen
for scenario in tests:
    scenario.df.loc["Anteil Peilungen mit Antwort"] = scenario.df.loc["Anzahl Peilungen mit Antwort"] / scenario.df.loc["Anzahl Peilungen Insgesamt"]

```

```python
# Misc

FIG_SIZE = np.array([8, 5])

all_angles = set([int(a) for scenario in tests for a in scenario.df.columns])
ANGLES = sorted(list(all_angles))
AANGLES = [-180] + ANGLES

def make_x_angle(ax: plt.Axes):
    ax.set_xticks(AANGLES)
    ax.set_xlim(-180, 180)

def make_y_angle(ax: plt.Axes):
    ax.set_yticks(AANGLES)
    ax.set_ylim(-180, 180)

def pscatter(ax: plt.Axes, Z: pd.Series, **kwargs):
    ax.scatter(Z.index, Z.values, **kwargs)

plt.Axes.make_x_angle = make_x_angle
plt.Axes.make_y_angle = make_y_angle
plt.Axes.pscatter = pscatter
```

```python
# Anteil Peilungen mit Antwort

fig, ax = plt.subplots(figsize=FIG_SIZE)
ax.set_xlabel("α in °")
ax.set_ylabel("Antwortrate")

ax.set_ylim(0, 1)
ax.make_x_angle()
ax.grid(True, which="major", axis="both")

for scenario in tests:
    ax.plot(scenario.df.loc["Anteil Peilungen mit Antwort"], label=scenario.name(), color=scenario.color)

ax.legend() #loc="lower left"
fig.savefig("plots/Antwortrate.pdf", bbox_inches="tight")
```

```python
# MUSIC Summe Peaks vs Alpha

fig, ax = plt.subplots(figsize=FIG_SIZE)
ax.set_xlabel("α in °")
ax.set_ylabel("Peak MUSIC-Summe in °")

ax.make_x_angle()
ax.make_y_angle()
ax.grid(True, which="major", axis="both")

ax.plot(AANGLES, AANGLES, color='lightgray')

for scenario in tests:
    ax.plot(scenario.df.columns, scenario.df.loc["MUSIC Summe - Peak"], label=scenario.name(), color=scenario.color)

ax.legend()
fig.savefig("plots/MUSIC_Summe_Peaks_vs_Alpha.pdf", bbox_inches="tight")
```

```python
# PDOA Mittelwert/Median vs Alpha
def pdoa(m: Literal['Mittelwert', 'Median', 'Standardabweichung', 'RMSE']):
    ucaRange01 = { -30, 0, 30, 60, 90, 120, 150}
    ucaRange12 = ({-180, -150, -120, -90}, {90, 120, 150, 180})
    ucaRange02 = {-150, -120, -90, -60, -30, 0, 30}

    fig = plt.Figure(figsize=FIG_SIZE * 2)
    gs = fig.add_gridspec(2, 4)
    gs.update(wspace=0.4, hspace=0.3)
    ax_tl = fig.add_subplot(gs[0, 0:2])
    ax_tr = fig.add_subplot(gs[0, 2:4])
    ax_bc = fig.add_subplot(gs[1, 1:3])
    axs = [ax_tl, ax_tr, ax_bc]

    # fig, axs = plt.subplots(2, 2, figsize=FIG_SIZE * 2)
    # axs: list[plt.Axes] = axs.flatten()


    for ax in axs:
        ax.sharex(axs[0])
        ax.sharey(axs[0])
        ax.set_xlabel("α in °")
        ax.make_x_angle()
        if m in ('Mittelwert', 'Median'):
            ax.make_y_angle()
            ax.plot(AANGLES, AANGLES, color='lightgray')
        ax.grid(True, which="major", axis="both")

    axs[0].set_ylabel(f"{m} PDOA 0-1 in °")
    axs[1].set_ylabel(f"{m} PDOA 1-2 in °")
    axs[2].set_ylabel(f"{m} PDOA 0-2 in °")
    axs[0].set_title("PDOA 0-1")
    axs[1].set_title("PDOA 1-2")
    axs[2].set_title("PDOA 0-2")

    for scenario in tests:
        if scenario.type == 'ULA':
            axs[0].plot(scenario.df.loc[f"PDOA01 - {m}"], label=scenario.name(), color=scenario.color)
            axs[1].plot(scenario.df.loc[f"PDOA12 - {m}"], label=scenario.name(), color=scenario.color)
            axs[2].plot(scenario.df.loc[f"PDOA02 - {m}"], label=scenario.name(), color=scenario.color)
        else:
            axs[0].plot(scenario.df.loc[f"PDOA01 - {m}"].where(scenario.df.columns.isin(ucaRange01   ), np.nan), label=scenario.name(), color=scenario.color)
            axs[1].plot(scenario.df.loc[f"PDOA12 - {m}"].where(scenario.df.columns.isin(ucaRange12[0]), np.nan), label=scenario.name(), color=scenario.color)
            axs[1].plot(scenario.df.loc[f"PDOA12 - {m}"].where(scenario.df.columns.isin(ucaRange12[1]), np.nan), label=scenario.name(), color=scenario.color)
            axs[2].plot(scenario.df.loc[f"PDOA02 - {m}"].where(scenario.df.columns.isin(ucaRange02   ), np.nan), label=scenario.name(), color=scenario.color)

    axs[0].legend()
    return fig

pdoa('Mittelwert')        .savefig("plots/PDOA_Mittelwert_vs_Alpha.pdf", bbox_inches="tight")
pdoa('Median')            .savefig("plots/PDOA_Median_vs_Alpha.pdf", bbox_inches="tight")
pdoa('Standardabweichung').savefig("plots/PDOA_Standardabweichung_vs_Alpha.pdf", bbox_inches="tight")
pdoa('RMSE')              .savefig("plots/PDOA_RMSE_vs_Alpha.pdf", bbox_inches="tight")
```
