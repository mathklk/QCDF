```python
# Misc

all_angles = set([int(a) for scenario in tests for a in scenario.df.columns])
ANGLES = sorted(list(all_angles))
AANGLES180 = [-180] + ANGLES
AANGLES90 = [x for x in ANGLES if abs(x) <= 90]

# @axis dector to adds method to plt.Axes class
def axis(func):
    setattr(plt.Axes, func.__name__, func)
    return func

@axis
def make_x_angle(ax: plt.Axes, angles: List[int]):
    ax.set_xticks(angles)
    ax.set_xlim(min(angles), max(angles))

@axis
def make_y_angle(ax: plt.Axes, angles: List[int]):
    ax.set_yticks(angles)
    ax.set_ylim(min(angles), max(angles))

@axis
def pscatter(ax: plt.Axes, Z: pd.Series, **kwargs):
    ax.scatter(Z.index, Z.values, **kwargs)

@axis
def wrapplot(ax: plt.Axes, Z: pd.Series, low = -180, high = 180, **kwargs):
    pieces: Iterable[pd.Series] = [pd.Series(dtype=Z.dtype)]
    wrapRange = high - low
    prev = None
    for i in range(len(Z)):
        x = Z.index[i]
        y = Z.values[i]
        if np.isnan(y):
            continue
        if prev is None:
            pieces[-1].loc[x] = y
        else:
            if abs(y - prev[1]) <= wrapRange / 2:
                pieces[-1].loc[x] = y
            else:
                pieces[-1].loc[x] = y - wrapRange * np.sign(y - prev[1])
                pieces.append(pd.Series(dtype=Z.dtype))
                pieces[-1].loc[prev[0]] = prev[1] + wrapRange * np.sign(y - prev[1])
                pieces[-1].loc[x] = y
        prev = (x, y)
    for piece in pieces:
        ax.plot(piece.index, piece.values, **kwargs)
        if 'label' in kwargs:
            del kwargs['label']  # only add label once, or it will appear multiple times in the legend
```
