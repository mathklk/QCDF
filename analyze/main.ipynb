```python
import pandas as pd
import io
import dataclasses
from typing import *
import matplotlib.pyplot as plt
```

```python
# load previous cell as text

@dataclasses.dataclass
class Scenario:
    type: str
    distance: float
    data: str
    color: str

    df: pd.DataFrame = None # type: ignore


tests = [
    Scenario('ULA', 3.5, ula3_5, '#4472c4'),
    Scenario('ULA', 20,  ula20 , '#9bc2e6'),
    Scenario('ULA', 100, ula100, '#9bc2e6'),
    Scenario('UCA', 20,  uca20 , '#ed7d31')
]

# String Copy/Paste interpretieren

for i, scenario in enumerate(tests):
    df = pd.read_csv(io.StringIO(scenario.data), sep="\t", header=None)
    df.index = [
        "Anzahl Peilungen Insgesamt",
        "Anzahl Peilungen mit Antwort",
        "MUSIC Peaks Separat - Mittelwert",
        "MUSIC Peaks Separat - Median",
        "MUSIC Peaks Separat - Standardabweichung",
        "MUSIC Peaks Separat - RMSE",
        "MUSIC Summe - Peak",
        "MUSIC Summe - Qualität",
        "PDOA01 - Mittelwert",
        "PDOA01 - Median",
        "PDOA01 - Standardabweichung",
        "PDOA01 - RMSE",
        "PDOA12 - Mittelwert",
        "PDOA12 - Median",
        "PDOA12 - Standardabweichung",
        "PDOA12 - RMSE",
        "PDOA02 - Mittelwert",
        "PDOA02 - Median",
        "PDOA02 - Standardabweichung",
        "PDOA02 - RMSE",
        "PDOA Mittel - Mittelwert",
        "PDOA Mittel - Median",
        "PDOA Mittel - Standardabweichung",
        "PDOA Mittel - RMSE"
    ]
    if scenario.type == 'ULA':
        df.columns = [-90, -60, -30, 0, 30, 60, 90]
    else:
        df.columns = [-150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180]
    tests[i].df = df

# Automatische, "Statische" Berechnungen
for scenario in tests:
    scenario.df.loc["Anteil Peilungen mit Antwort"] = scenario.df.loc["Anzahl Peilungen mit Antwort"] / scenario.df.loc["Anzahl Peilungen Insgesamt"]

```

```python
# Plotte "Antel Peilungen mit Antwort"
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title("Anteil Peilungen mit Antwort")
ax.set_xlabel("Alpha / °")
ax.set_ylabel("Antwortrate")
ax.grid()
for scenario in tests:
    ax.plot(scenario.df.columns, scenario.df.loc["Anteil Peilungen mit Antwort"], label=f"{scenario.type} {scenario.distance}m", color=scenario.color)
```
