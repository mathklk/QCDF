```python
def plotScenario(scenario: Scenario, ax: plt.Axes):
    # fig, ax = plt.subplots(figsize=FIG_SIZE)
    ax.set_title("MUSIC " + scenario.name())
    df = scenario.df
    angles = df.index.astype(float)
    frequencies = df.columns.astype(float)

    Z_norm = (df.values - df.values.min(axis=1, keepdims=True)) / (df.values.max(axis=1, keepdims=True) - df.values.min(axis=1, keepdims=True))

    for i, angle in enumerate(angles):
        y_offset = i
        overlap = .9
        spectrum = Z_norm[i] * overlap
        ax.fill_between(
            frequencies, 
            y_offset, 
            y_offset + spectrum, 
            alpha=0.7, 
            color=scenario.color, 
            edgecolor='black', 
            linewidth=1, 
            zorder=100-i
        )
        ax.vlines(
            angle, 
            y_offset, 
            y_offset + overlap, 
            colors='black', 
            linestyles='--', 
            linewidth=1.0, 
            zorder=101-i
        )

    yticks = range(0, len(angles), max(1, len(angles)//10))
    ax.set_yticks(yticks)
    ax.set_yticklabels([f"{int(angles[i])}°" for i in yticks])
    xticks = list(df.index.astype(float))
    if scenario.type == 'UCA':
        xticks = [-180] + xticks
    ax.set_xticks(xticks)

    ax.set_xlabel('Pseudospektrum Winkel in °', fontsize=12)
    ax.set_ylabel('α in °', fontsize=12)
    # ax.set_title(f'MUSIC Summen Spektrum - {scenario.name()}')
    # ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)

    # _min, _max = df.values.min(), df.values.max()
    # print(f"Scenario {scenario.name()} min: {_min}, max: {_max}")

ula_fig, ula_axs = plt.subplots(3, figsize=FIG_SIZE*[1,3])
# set vspace higher
ula_fig.subplots_adjust(hspace=0.3)

for (scenario, ax) in zip([test for test in tests if test.type == 'ULA'], ula_axs):
    plotScenario(scenario, ax)

uca_fig, uca_axs = plt.subplots(figsize=FIG_SIZE)
plotScenario([test for test in tests if test.type == 'UCA'][0], uca_axs)

ula_fig.savefig('plots/MUSIC_Summe_Ridges_ULA.pdf', bbox_inches='tight')
uca_fig.savefig('plots/MUSIC_Summe_Ridges_UCA.pdf', bbox_inches='tight')
```
